```python
import os
import cv2
import random
import numpy as np
from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader, Dataset
from torchvision.transforms import transforms
from torchvision.datasets import ImageFolder
from torchvision.models import resnet34
import torch
import torch.nn as nn
import shutil
##edit the pics 

a_transform = transforms.Compose([
    transforms.RandomResizedCrop(size=224),
#    transforms.RandomHorizontalFlip(),
#    transforms.RandomAffine(degrees = (-90.,90.), translate=(0.1,0.1), scale = (0.9,1.1)),
    transforms.ToTensor(),
    transforms.RandomErasing(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])])

e_transform = transforms.Compose([
    transforms.Resize(size=224),
    transforms.CenterCrop(size=224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])])
        
        
train_dataset = ImageFolder(f"{DATASETDIR}/Train", transform=a_transform)
val_dataset = ImageFolder(f"{DATASETDIR}/Validation", transform=e_transform)
test_dataset = ImageFolder(f"{DATASETDIR}/Test", transform=e_transform)
```

```python
model = resnet34(pretrained=True)
model.fc = nn.Sequential(
    nn.Dropout(0.5),
    nn.Linear(in_features=512, out_features=len(val_dataset.classes), bias=True)
)
model = model.to("cuda")
```

```python
loss_func = torch.nn.CrossEntropyLoss()
```

```python
optimizer = torch.optim.SGD(model.parameters(), lr=1e-3, momentum=0.9)
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=3, gamma=0.9)
```

```python
def training(model, dataset, optimizer, loss_func):
    model.train()
  
    losses = []
    for img, target in DataLoader(dataset, shuffle=True, batch_size=32, num_workers=4):
        img = img.to("cuda")
        target = target.to("cuda")
        model = model.to("cuda")
        pred = model(img)
        loss = loss_func(pred, target)
        losses.append(loss.item())

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    
    return np.mean(losses)
    
```

```python
def evaluation(model, dataset, loss_func):
    model.eval()
  
    losses = []
    correct = 0
    with torch.no_grad():
        for img, target in DataLoader(dataset, shuffle=False, batch_size=64, num_workers=4):
            img = img.to("cuda")
            target = target.to("cuda")
            pred = model(img)
            loss = loss_func(pred, target)
            losses.append(loss.item())

            predict = pred.argmax(dim=1)
            correct += (predict == target).sum().item()
    return np.mean(losses), correct / len(dataset)
```

```python
best_loss = 1e+10
best_state = None
earlystop_counter = 0
for epoch in range(15):
    train_loss = training(model, train_dataset, optimizer, loss_func)
    val_loss, val_acc = evaluation(model, val_dataset, loss_func)
    if val_loss < best_loss:
        earlystop_counter = 0
        best_loss = val_loss
        best_state = model.cpu().state_dict()
        model.cuda()
    else:
        earlystop_counter += 1
    if earlystop_counter > 5:
        break
    scheduler.step()
    print(f"epoch {str(epoch).zfill(2)}\ttrain_loss: {train_loss}\tval_loss: {val_loss}\tval_acc: {val_acc}")
```

```python
model.load_state_dict(best_state)
test_loss, test_acc = evaluation(model, test_dataset, loss_func)
print(f"Acc in Test：{test_acc}")
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def create_confusion_matrix(cm, labels):
    sns.set()
    
    df = pd.DataFrame(cm)
    df.index = labels
    df.columns = labels

    fig, ax = plt.subplots(figsize=(15, 10))
    sns.heatmap(df, annot=True, fmt="d", linewidths=.5, ax=ax, cmap="YlGnBu")

def plot_confusion_matrix(model, dataset):
    images_so_far = 0
    fig = plt.figure()

    targets = np.array([])
    preds = np.array([])

    model.to("cuda")
    model.eval()
    with torch.no_grad():
        for img, target in DataLoader(dataset, shuffle=False, batch_size=64, num_workers=4):
            img = img.to("cuda")
            target = target.to("cuda")
            pred = model(img)
            pred = pred.argmax(dim=1)

            targets = np.append(targets, target.cpu().data.numpy())
            preds = np.append(preds, pred.cpu().numpy())
    
    cm = confusion_matrix(targets, preds)
    create_confusion_matrix(cm, dataset.classes)
```

```python
plot_confusion_matrix(model, test_dataset)
```

```python
def get_result(model, dataset):
    model.to("cuda")
    model.eval()

    all_labels = np.array([])
    all_preds = np.array([])
    with torch.no_grad():
        for img, target in DataLoader(dataset, shuffle=False, batch_size=64, num_workers=8):
            img = img.to("cuda")
            target = target.to("cuda")

            pred = model(img)
            pred_label = pred.argmax(1)
            
            all_labels = np.append(all_labels, target.cpu().data.numpy())
            all_preds = np.append(all_preds, pred_label.cpu().numpy())
            
    return all_labels.astype(int).tolist(), all_preds.astype(int).tolist()
```
